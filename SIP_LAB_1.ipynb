```python
#import numpy,matplotlib.pyplot,signal form scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sg
import warnings
warnings.filterwarnings('ignore')
```

```python
t=np.arange(0,1,0.01)
f=5
x=2*np.sin(2*np.pi*f*t)
```

```python
plt.plot(t,np.round(x))
plt.title("Plot of the sine wave - Discrete Amplitude")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
plt.stem(t,np.round(x))
plt.title("Plot of the sine wave - Discrete Amplitude")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

# UNIT STEP

```python
#unit step
t = np.arange(-10,10,0.5)
un = []
for i in t:
  if i >= 0:
    un.append(1)
  else:
    un.append(0)
```

```python
plt.stem(t,un)
plt.title("Unit Step")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

#UNIT RAMP

```python
#unit ramp
t = np.arange(-10,10,0.5)
rn = []
for i in t:
  if i >= 0:
    rn.append(i)
  else:
    rn.append(0)
```

```python
plt.stem(t,rn)
plt.title("Unit Ramp")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

# UNIT IMPULSE

```python
#unit impulse
t = np.arange(-10,10,0.5)
imp = []
for i in t:
  if i == 0:
    imp.append(1)
  else:
    imp.append(0)
```

```python
plt.stem(t,imp)
plt.title("Unit Impluse")
plt.grid()
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

#TIME SHIFTING

```python
#shift

n = np.arange(-10, 10, 1)
u = np.where(n>=0, n, 0)

plt.subplot(2, 2, 1)
plt.plot(n, u)
plt.subplot(2, 2, 2)
plt.plot(-n, u)
plt.subplot(2, 2, 3)
plt.plot(n + 5, u)
plt.subplot(2, 2, 4)
plt.plot(-n + 5, u)
```

# TIME REVERSAL

```python
#Time Reversal
x= np.arange(-1,1.5,0.5)
y=x[::-1]
plt.subplot(1,2,1)
plt.stem(x,x)
plt.subplot(1,2,2)
plt.stem(x,y)
```

#TIME SHIFTING

```python
#Time Shifting
x= np.arange(-1,1.5,0.5)
y=x-0.2
plt.subplot(1,2,1)
plt.stem(x,x)
plt.subplot(1,2,2)
plt.stem(x,y)
print(y)
```

# TIME SCALING

```python
#Time Scaling
x= np.arange(-1,1.5,0.5)
y=2*x
plt.subplot(1,2,1)
plt.stem(x,x)
plt.subplot(1,2,2)
plt.stem(x,y)
```
