## Grey Scale Resolution

```python
img = cv2.imread("cameraman.tif",0)
```

```python
np.max(img)
```

```python
[m,n]= img.shape
```

```python
k = int(input("Enter no. of required bits for the new image = "))
```

```python
original = 256 #2^8
new = 2**k
Const = new/original
Const
```

```python
img2 = np.zeros((m,n),dtype= np.int)

for i in range(m):
  for j in range(n):

    img2[i,j] = Const*img[i,j]
```

```python
new_img2 = Const*img
new_img2
```

```python
plt.imshow(img2, cmap ='gray')
```

```python
plt.subplot(2,2,1)
plt.imshow(img,cmap ='gray')
plt.title('Original')
plt.subplot(2,2,2)
plt.imshow(img2,cmap ='gray')
plt.title('Reduced Grayscale Resolution')
```
