```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import signal
```

```python
img = cv2.imread("/content/Fig1016(a)(building_original).tif",0)
m,n = img.shape
```

Horizontal Edge Detection

```python
def horizontal(img):
  sobel_h = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
  img_h = np.zeros([m,n])

  for i in range(1, m-1):
    for j in range(1,n-1):
      img_h[i,j] = np.multiply(img[i-1:i+2,j-1:j+2],sobel_h).sum() 

  plt.title("Horizontal Edges")
  plt.imshow(img_h,cmap="gray", vmin=0, vmax=255)

  return(img_h)
```

Vertical Edge Detection

```python
def vertical(img):
  sobel_v = np.array([[-1,0,1], [-2,0,2], [-1,0,1]])
  img_v = np.zeros([m,n])

  for i in range(1, m-1):
    for j in range(1,n-1):
      img_v[i,j] = np.multiply(img[i-1:i+2,j-1:j+2],sobel_v).sum() 

  plt.title("Vertical Edges")
  plt.imshow(img_v,cmap="gray", vmin=0, vmax=255)

  return(img_v)
```

Diagonal Edge Detection

```python
def diagonal(img):
  sobel_d = np.array([[0,1,2], [-1,0,1], [-2,-1,0]])
  img_d = np.zeros([m,n])

  for i in range(1, m-1):
    for j in range(1,n-1):
      img_d[i,j] = np.multiply(img[i-1:i+2,j-1:j+2],sobel_d).sum() 

  plt.title("Diagonal Edges")
  plt.imshow(img_d,cmap="gray", vmin=0, vmax=255)

  return(img_d)
```

Image Low Pass Filter 

```python
img_lpf= np.zeros([m,n], dtype=int)
mask = 6
a = mask//2

for i in range(a,m-a):
    for j in range(a,n-a):
        temp= img[i-a:i+a+1,j-a:j+a+1] 
        
        constant= np.sum(temp)
        img_lpf[i,j]= constant//mask**2
```

Plotting the Graphs

```python
#original and lpf image
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img,cmap="gray", vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.title("Original LPF Image")
plt.imshow(img_lpf,cmap="gray", vmin=0, vmax=255)

#horizontal edges
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
img_h = horizontal(img)
plt.subplot(1,2,2)
img_lpf_h = horizontal(img_lpf)

#vertical edges
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
img_v = vertical(img)
plt.subplot(1,2,2)
img_lpf_v = vertical(img_lpf)

#diagonal edges
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
img_d = diagonal(img)
plt.subplot(1,2,2)
img_lpf_d = diagonal(img_lpf)

#all edges
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
img_hvd = img_h + img_v + img_d
plt.title("All Edges Image")
plt.imshow(img_hvd,cmap="gray", vmin=0, vmax=255)
plt.subplot(1,2,2)
img_lpf_hvd = img_lpf_h + img_lpf_v + img_lpf_d
plt.title("All Edges Image")
plt.imshow(img_lpf_hvd,cmap="gray", vmin=0, vmax=255)
```
