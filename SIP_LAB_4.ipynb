```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
```

```python
img1 = cv2.imread("/content/cameraman.tif")
cv2.imwrite("Image_new.tif",img1)
plt.imshow(img1)
```

```python
type(img1)
```

```python
img1 = img1[:,:,0]
```

```python
L = np.max(img1)
L
```

## NEGATIVE

```python
img_neg = L - img1
print(img_neg)
print(img_neg.shape)
plt.imshow(img_neg, cmap = 'gray')
```

## THRESHOLD

```python
threshold = int(input("Enter the threshold: "))
img3 = np.where(img1 > threshold,255,0)
print(img3)
print(img3.shape)
plt.imshow(img3, cmap = 'gray')
```

## GRAY LEVEL SLICING (without background)

```python
a = int(input("Enter start range for gray level slicing: "))
b = int(input("Enter end range for gray level slicing: "))

img4 = np.where(np.logical_and(img1>a,img1<b), 255,0)
print(img4)
print(img4.shape)
plt.imshow(img4, cmap = 'gray')
```

## GRAY LEVEL SLICING (with background)

```python
a = int(input("Enter start range for gray level slicing: "))
b = int(input("Enter end range for gray level slicing: "))

img4 = np.where(np.logical_and(img1>a,img1<b), 255,img1)
print(img4)
print(img4.shape)
plt.imshow(img4, cmap = 'gray')
```

## LOW PASS FILTERING

```python
m,n = img1.shape
```

```python
img_gen= np.zeros([m,n], dtype=int)
b= int(input("Enter size of the mask: "))
a=b//2

for i in range(a,m-a):
    for j in range(a,n-a):
        temp= img1[i-a:i+a+1,j-a:j+a+1] 
        
        constant= np.sum(temp)
        img_gen[i,j]= constant//b**2
```

```python
plt.imshow(img_gen,cmap="gray" )
```
