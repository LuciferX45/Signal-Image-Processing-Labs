```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/Fig0911(a)(noisy_fingerprint).tif",0)
m,n = img.shape 
plt.imshow(img,cmap = "gray")
```

```python
def dilation(img):
  
  imgDilate= np.zeros((m,n), dtype= np.uint8)
  SED = np.array([[1,1,1], [1,1,1], [1,1,1]])
  constantD = 1

  for i in range(constantD, m-constantD):
    for j in range(constantD, n-constantD):
      temp= img[i-constantD:i+constantD+1, j-constantD:j+constantD+1]
      product= temp*SED
      imgDilate[i,j]= np.max(product)

  return(imgDilate)
```

```python
def erosion(img):

  SE= np.ones((3,3), dtype=np.uint8) 
  constant = 1

  imgErode= np.zeros((m,n), dtype= np.uint8)

  for i in range(constant, m-constant):
    for j in range(constant, n-constant):
      temp= img[i-constant:i+constant+1, j-constant:j+constant+1]
      product= temp*SE
      imgErode[i,j]= np.min(product)
  
  return(imgErode)
```

```python
plt.figure(figsize=(20,20))

plt.subplot(1,5,1)
plt.imshow(img,cmap = "gray")
plt.title("Original Image")

#opening
plt.subplot(1,5,2)
imgE = erosion(img)
plt.imshow(imgE,cmap = "gray")
plt.title("Eroded Image for Opening")

plt.subplot(1,5,3)
imgO = dilation(imgE)
plt.imshow(imgO, cmap = "gray")
plt.title("Image after Opening")

#closing
plt.subplot(1,5,4)
imgD = dilation(imgO)
plt.imshow(imgD, cmap = "gray")
plt.title("Dilated Image for Closing")

plt.subplot(1,5,5)
imgC = erosion(imgD)
plt.imshow(imgC, cmap = "gray")
plt.title("Image after Closing")
```

#**CONCLUSION**
We learnt how to apply erosion and dilation and perform opening and closing of an image.
