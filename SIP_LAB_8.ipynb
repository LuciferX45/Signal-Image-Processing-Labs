```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/content/Fig0905(a)(wirebond-mask).tif",0)
m1,n1 = img1.shape 
```

```python
img2 = cv2.imread("/content/Fig0907(a)(text_gaps_1_and_2_pixels).tif",0)
m2,n2 = img2.shape
```

## EROSION

```python
def erosion(img,k):

  SE= np.ones((k,k), dtype=np.uint8) 
  constant= (k-1)//2

  imgErode= np.zeros((m1,n1), dtype= np.uint8)

  for i in range(constant, m1-constant):
    for j in range(constant, n1-constant):
      temp= img[i-constant:i+constant+1, j-constant:j+constant+1]
      product= temp*SE
      imgErode[i,j]= np.min(product)
  
  return(imgErode)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(img1, cmap="gray")
plt.title("Original")
plt.subplot(1,4,2)
plt.imshow(erosion(img1,11), cmap="gray")
plt.title("Eroded with SE of size {} X {}".format(11,11))
plt.subplot(1,4,3)
plt.imshow(erosion(img1,15), cmap="gray")
plt.title("Eroded with SE of size {} X {}".format(15,15))
plt.subplot(1,4,4)
plt.imshow(erosion(img1,45), cmap="gray")
plt.title("Eroded with SE of size {} X {}".format(45,45))
```
